```python
import pandas as pd
from datetime import *
import yaml
```

```python
# get the path to the config.yml file
path = os.getcwd()
configfile_2 = 'config.yml'
file_path = os.path.join(path,configfile_2)
```

```python
# read in the config file
with open(file_path, 'r') as file:
    data_config = yaml.safe_load(file)
```

```python
# True or False
bought_input_data=False
```

```python
# Load the input dataset
country_data = data_config['uk_weather_data'] # change for germany and netherlands

weather_data=data_config['input_data'] + country_data
data = pd.read_csv(weather_data)
data.tail()
```

```python
if bought_input_data==True:
    df=data.copy()
    df['dt_iso']=df['dt']
    df['dt_iso'] = df['dt_iso'].apply(lambda x: pd.to_datetime(str(x).split(' ')[0], format='%Y-%m-%d'))
    df = df.groupby('dt_iso').aggregate('mean')
    df['year-week'] = pd.Series(df.index).apply(lambda x: str(x.year)+"-"+str(x.week).zfill(2)).values
    print(df)

else:
    df=data.copy()
    df['dt_iso']=df['dt']

    # create new columns for mean, max and avg and combine them all

    df['dt_iso'] = df['dt_iso'].apply(lambda x: pd.to_datetime(str(x).split(' ')[0], format='%Y-%m-%d'))
    df_min=df.groupby('dt_iso').min().rename(columns = {'temp':'temp_min'}).drop(['dt'], axis=1)
    df_max=df.groupby('dt_iso').max().rename(columns = {'temp':'temp_max'}).drop(['dt'], axis=1)
    df_mean=df.groupby('dt_iso').aggregate('mean')
    df=pd.concat([df_mean,df_min,df_max],axis=1)

    # extract the week and month of the year
    
    df['year-week'] = pd.Series(df.index).apply(lambda x: str(x.year)+"-"+str(x.week).zfill(2)).values
    df["week_ending"] = pd.Series(df.index).apply(lambda x: x - timedelta(days=x.dayofweek)).values
    print(df.tail(50))

```
